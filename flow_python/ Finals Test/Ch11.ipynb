## 11.4 하나의 차트에 여러 개의 데이터를 그려보자

- linear: $y_1 = x $
- quadratic : $y_2 = x ^ 2$   
- cubic: $y_3 = x ^ 3$

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 101)
# x = np.arrange(-5, 5.01, 0.1)
y1 = -x
y2 = x ** 2
y3 = x ** 3

plt.plot(x, y1, 'k.-', label ='linear')
plt.plot(x, y2, 'r^:', label = 'quadratic')
plt.plot(x, y3, 'b*--', label = 'cublic')

plt.xlim([-2, 2])
plt.ylim([-10, 10])
plt.legend()
plt.show()
````

**LAB11-2** 삼각함수의 기본인 사인 그래프 그리기

````python
# import package

# 각도 0-360
theta = np.arange(0, 360)

# 사인 값 만들기
y1 = np.sin(2*np.pi*theta / 360)
y2 = np.cos(np.radians(theta))
# np.sin(np.radians(theta))

# 그리기
plt.plot(theta, y1, 'b')
plt.plot(theta, y2, 'y')

plt.title('sin')
plt.xlabel('theta')
plt.ylabel('radians(theta)')
plt.grid()
plt.show()
````

## 11.7 히스토그램으로 자료의 분포를 한눈에 살펴보자

- **히스토그램histogram**은 주어진 자료를 몇 개의 구간으로 나누고 각 구간의 *도수frequency*를 조사하여 나타낸 막대 그래프이다.

```
books = [1, 6, 2, 3, 1, 2, 0, 2]
```

![322-2.JPG](322-2.JPG)

````python
books = [1, 6, 2, 3, 1, 2, 0, 2]

sorted(books)
````

````python
계급  : 계급값 : 빈도수(frequency)
[0,1): 0.5 : 1
[1,2): 1.5 : 2
[2,3): 2.5 : 3
[3,4): 3.5 : 1
[4,5): 4.5 : 0
[5,6): 5.5 : 1
````

- 히스토그램 그리는 순서
1) 최대 값과 최소 값을 구한다.
2) 최대 값과 최소 값을 정해진 구간(bin)으로 나눈다.
3) 각 빈에 있는 샘플의 수(도수)를 센다.

````python
books = [1, 6, 2, 3, 1, 2, 0, 2]
plt.hist(books, bins = 6)
plt.show()
````
