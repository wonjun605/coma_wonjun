## 11.2
넘파이의 sin(), cos() 함수를 이용하여 주기 함수를 그려보자. 여섯 주기까지 그린다. 사인 함수의 진폭은 $x$ 값이 커질수록 커지게 한다. 코사인 함수의 진폭은 20이다.


````python
x = np.arange(360 * 6)
x = np.radians(x)

y1 = x * np.cos(x)
y2 = 20 * np.cos(x)

plt.figure(figsize = (6, 3))
plt.plot(x, y1, 'r-', label = 'sin')
plt.plot(x, y2, 'b--', label = 'cos')
plt.show()
````
